```python
def manhattan(X,Y):
    return abs(X[0] - Y[0]) + abs(X[1] - Y[1])
```

```python
import numpy as np

# Load the dataset as a numpy array
X = np.array([[2,3],  
    [2,4],
    [6,5],
    [6,3] ])

from scipy.spatial import distance
import pandas as pd
a = X[0]
b = X[1]
c = X[2]
d = X[3]

dist = {'A': [0, manhattan(a,b), manhattan(a,c), manhattan(a, d)],
   'B': [manhattan(a,b), 0, manhattan(b,c), manhattan(b,d)],
   'C': [manhattan(a, c),manhattan(b,c) , 0, manhattan(c,d)],
   'D': [manhattan(a, d),manhattan(b,d) ,manhattan(c,d) , 0]
        }

df = pd.DataFrame(dist, columns= ['A', 'B', 'C', 'D'])
print(df)
```

```python
# Load the dataset as a numpy array
X = np.array([[2,3],  
    [2,4],
    [6,5],
    [6,3] ])
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage  
from matplotlib import pyplot as plt

# create a dendogram
linked = linkage(X, 'single')

labelList = X

plt.figure(figsize=(8,6))  
dendrogram(linked, orientation='top', labels=labelList,
            distance_sort='descending', show_leaf_counts=True)
plt.show() 
```

```python
from sklearn.cluster import AgglomerativeClustering

# create an object of Agglomerative clustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='manhattan', linkage='complete')  

# train and predict
cluster.fit_predict(X)
```

## Dendogram on Iris datset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load iris dataset
from sklearn.datasets import load_iris

data, target = load_iris(return_X_y=True)
```

```python
from sklearn.preprocessing import normalize

data_scaled = normalize(data)
data_scaled = pd.DataFrame(data_scaled)
data_scaled.head()
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  

dend = shc.dendrogram(shc.linkage(data, method='complete')) 

# Affinity = {euclidean, l1, l2, manhattan, cosine}
# Linkage = {ward, complete, average, single}
# Ward Linkage = sum of squared differences within all clusters
```

## Agglomerative Clustering on Iris dataset

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')  
cluster.fit_predict(data_scaled)
```
